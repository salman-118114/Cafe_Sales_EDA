```python
import pandas as pd
import numpy as np
```

Lets clean the data set

```python
df = original_df.copy()
```

from this column i can see these are data inconsistency lets fix that first

```python
airline_fix = {
    'Vistara Premium economy' : 'Vistara',
    'Jet Airways Business' : 'Jet Airways',
    'Multiple carriers Premium economy':'Multiple carriers'
}

df['Airline'] = df['Airline'].replace(airline_fix)
```

```python
df['Airline'].unique()
```

```python
df['Airline'].nunique()
```

Data journey column cleaning


```python
#first convert journey dates in seperate columns for day, month, year
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'], dayfirst=True)
df['Day_of_journey'] = df['Date_of_Journey'].dt.day
df['month_of_journey'] = df['Date_of_Journey'].dt.month
df['year_of_journey'] = df['Date_of_Journey'].dt.year
df['day_of_week_journey'] = df['Date_of_Journey'].dt.dayofweek
df['is_weekend_journey'] = df['day_of_week_journey'].isin([5,6]).astype(int)
```

```python
df.head()
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
```

date of journey converted to datetime formate and seperated day, month, year , day of week and weekend or not also dropped the orginal date of journey column

```python
### now cleaning source column
df['Source'].isna().sum()
```

```python
df['Source'].unique()
```

source column looks good

```python
# destination column cleaning
df['Destination'].isna().sum()
```

```python
df['Destination'].unique()
```

Destination column also looks good

```python
# next lets focus on departure time column (dep_time)
df['Dep_Time'].isna().sum()
```

```python
df['Departure_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_minute'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1, inplace=True)
```

```python
df.head()
```

```python
## now lets clean arrival time column(arrival_time)
df['Arrival_Time'].isna().sum()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
df['Arrival_Time']
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minute'] = df['Arrival_Time'].str.split(':').str[1]
df.head()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
df.head()
```

```python
# lets claen duration column and convert it into minutes
df['Duration'].isna().sum()
```

```python
df['Duration'].info()
```

```python
df['Duration'] = pd.to_timedelta(df['Duration'])
```

```python
df.info()
```

```python
df['Duration']
```

```python
df['Duration_minutes'] = df['Duration'].dt.total_seconds() / 60
```

```python
df.head()
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
# removing route column as its not required
df.drop('Route',axis=1,inplace=True)
df.head()
```

 Cleaning Total stops its in categorical so convert into labels like 0,1,2... based on stops

```python
df['Total_Stops'].isnull().sum()
```

```python
df[df['Total_Stops'].isnull()]
```
